```python
import matplotlib.pyplot as plt

plt.figure(figsize = (26, 10))
plt.plot_date(alyx["Date"], 
              alyx["MSV"], 
              'b-',
              linewidth = 3,
             color = "black",
             label = "Alyx")
plt.plot_date(givenchy["Date"], 
              givenchy["MSV"], 
              'b-', 
              linewidth = 3,
             color = "red",
             label = "Givenchy")

plt.minorticks_on()
plt.xticks([4, 9, 14, 19, 24], ["6.5", "6.10", "6.15", "6.20", "6.25"])
# plt.yticks([50, 100, 200, 300, 400, 500, 550])
plt.xlabel("Date (June 2020)", labelpad = 25)
plt.ylabel("Mean Sales Value ($)", labelpad = 25)
plt.legend(fontsize = 19)
plt.axvline(x=14, 
            linestyle = "-.", 
            color = "red", 
            alpha = 0.7, 
            linewidth = 3)
plt.annotate("Date of Announcement",
             fontsize = 19,
             xy = (14.25, 515), 
             bbox = dict(facecolor = "none", edgecolor = "black", pad = 6))


# plt.title("Mean Sales Value", pad = 30)
```
