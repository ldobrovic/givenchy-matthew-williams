```python
import pandas as pd
import numpy as np

alyx = pd.DataFrame()

alyx = pd.read_csv("/Users/lukedobrovic/Data/Finished Projects/alyx/alyx.csv", index_col=0)
alyx.head()
```

```python
# Change sales with "hours ago" to "0 days ago"

counter = 0
for index, row in alyx.iterrows():
    if "hours" in row["Date"]:
        alyx.loc[counter, "Date"] = "0 days ago"
    counter +=1

alyx.head()
```

```python
# Drop all sales outside of the month of data collection

months = ["about 1 month ago", 
          "about 2 months ago", 
          "2 months ago", 
          "3 months ago", 
          "4 months ago", 
          "5 months ago", 
          "6 months ago",
          "7 months ago", 
          "8 months ago", 
          "9 months ago", 
          "10 months ago", 
          "11 months ago", 
          "12 months ago", 
          "about 1 year ago"]

indices = alyx.loc[alyx['Date'].isin(months)].index
alyx = alyx.drop(indices, axis=0)

alyx.tail()
```

```python
# Convert "Price" column from other to int

alyx['Price'] = alyx['Price'].replace('[\$\,\.]', '', regex=True).astype(int)
alyx.head()
```

```python
alyx["Price"].sum()
```

```python
# Create analytics DF with MSV, GSV, and TV by calculations on sales grouped by the day

analytics = pd.DataFrame()
analytics

MSV = round(alyx.groupby(["Date"])["Price"].mean(), 2)
analytics["MSV"] = MSV

GSV = round(alyx.groupby(["Date"])["Price"].sum(), 2)
analytics["GSV"] = GSV

TV = alyx.groupby(["Date"])["Date"].count()
analytics["TV"] = TV

analytics["Days Ago"] = analytics.index
analytics.reset_index(inplace=True, drop=True)

analytics
```

```python
# Gather Days Ago as int, use to sort the DF

counter = 0
for row in analytics.iterrows():
    phrase = analytics.loc[counter, "Days Ago"]
#     print(phrase)
    num = int(phrase.split(" ")[0])
    analytics.loc[counter, "Days Ago"] = num
    counter +=1
    
sorted = analytics.sort_values(by = ["Days Ago"], ascending = False)
```

```python
sorted
```

```python
import datetime

string_date = "2020-06-30"
origin = datetime.datetime.strptime(string_date, "%Y-%m-%d")

counter = 0
for row in sorted.iterrows():
    days_ago = int(sorted.loc[counter, "Days Ago"])
    delta = datetime.timedelta(days = days_ago)
    sorted.loc[counter, "Date"] = origin - delta
    counter +=1
    
# analytics.sort_values(by = "Days Ago", ascending = False)
sorted.reset_index(inplace=True, drop=True)
sorted.drop(sorted.head(1).index,inplace=True)

sorted.head()
```
